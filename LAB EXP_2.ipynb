```python
import pandas as pd

data = pd.read_csv("training_data.csv")
print(data)
```

```python
X = data.iloc[:, :-1]   # all columns except last
y = data.iloc[:, -1]    # last column (EnjoySport)

print(X)
print(y)
```

```python
num_attr = X.shape[1]   # number of attributes

S = ['0'] * num_attr
G = [['?'] * num_attr]

print("Initial S:", S)
print("Initial G:", G)
```

```python
for i in range(len(X)):
    x = list(X.iloc[i])
    label = y.iloc[i]

    print("\nTraining example:", x, "→", label)

    # POSITIVE example
    if label == 'Yes':
        for j in range(num_attr):
            if S[j] == '0':
                S[j] = x[j]
            elif S[j] != x[j]:
                S[j] = '?'

        G = [g for g in G if all(g[k] == '?' or g[k] == S[k]
                                 for k in range(num_attr))]

        print("Positive example")

    # NEGATIVE example
    else:
        new_G = []
        for g in G:
            for j in range(num_attr):
                if g[j] == '?' and S[j] != x[j]:
                    temp = g.copy()
                    temp[j] = S[j]
                    new_G.append(temp)

        G = new_G
        print("Negative example")

    print("S =", S)
    print("G =", G)
```

```python
print("\nFINAL RESULT")
print("Final Specific Hypothesis (S):", S)
print("Final General Hypotheses (G):", G)
```
